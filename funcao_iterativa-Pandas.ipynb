```python
# inves de usar o LOC OU NP.WHERE PARA TRATAR A VARIAVEL ALGUMAS VEZES É MELHOR TRATAR USANDO O df.apply como abaixo
import pandas as pd
dados = {
    'cpf_cnpj': ['11122233344', '22233344455', '33344455566', '44455566677', '55566677788',
                 '11222333444455', '22333444555566', '33444555666677', '44555666777788', '55666777888899'],
    'valor_devedor': [1500, 2200, 3500, 2800, 4200, 3900, 4700, 2900, 3800, 4800],
    'qtd_compras': [10, 8, 12, 7, 15, 9, 11, 13, 6, 14],
    'codigo': ['K', 'M', 'G', 'K', 'M', 'G', 'K', 'M', 'G', 'K'],
    'tipo_pessoa' : ['Física', 'Física', 'Física', 'Física', 'Física', 'Jurídica', 'Jurídica', 'Jurídica', 'Jurídica', 'Jurídica']
}

# Criando o DataFrame
df = pd.DataFrame(dados)
```

```python
# CRIA UMA FUNÇÃO ITERAVEL QUE PEGA TODOS OS ITENS INDIVIDUALMENTE E TRATA ELE RETORNANDO O TRATAMENTO QUE VOCÊ QUISER
def funcaoTratarLinhas(row):
    if row['tipo_pessoa'] == 'Física' and row['codigo'] == 'K':
        return 'Desconto de 50%'
    else:
        return 'Sem desconto'
    
df['fl_desconto'] = df.apply(funcaoTratarLinhas, axis=1)
display(df)
```

```python
# esse método que eu criei acima, estava sendo aplicado nas minhas ROTINAS, no entanto, percebi que ele demora MUIT0OOO, visto que precisa entrar em cada linha e tratar, sendo assim 
# Não recomendo usa-lo em grandes conjuntos de dados, especialmente na AWS, o ideal é usar o np.select


# COMO USAR:
    # 01. Defina as condições
    # 02. Defina os valores se as condições forem satisfeitas
    # 03. Defina o valor default se não for atingida todas soluções

import numpy as np
import pandas as pd

dados = {
    'renda_faturamento': [50000, 60000, 70000, 80000, 90000, 55000, 65000, 75000, 85000, 95000, 70000, 80000, 90000, 100000, 110000],
    'cpf_cnpj': [12345678901, 23456789012, 34567890123, 45678901234, 56789012345, 12345678902, 23456789013, 34567890124, 45678901235, 56789012346, 12345678903, 23456789014, 34567890125, 45678901236, 56789012347],
}

df = pd.DataFrame(dados)


condicao = [
    (df['renda_faturamento'] > 70000)
    ,(df['renda_faturamento'] >= 60000)
]

valores = ['Alto Nivel', 'Médio Nivel']


df['Nivel_cliente'] = np.select(condicao, valores, default='Baixo Nível')
display(df)
```
