```python
url = 'D:/credito.csv'
dt = pd.read_csv(url)
display(dt)
```

```python
# Verificação de valores nulos
nulos_por_coluna = dt.isnull().sum()

# Exibição dos resultados
print("Valores nulos por coluna:")
print(nulos_por_coluna)
```

```python
# Tratando dados 
dt.loc[dt['sexo'] == 'M', 'sexo'] = 'Masculino'
dt.loc[dt['sexo'] == 'F', 'sexo'] = 'Feminino' 

display(dt)


```

```python
dt['FX_QTD_PRODUTOS'] = np.where(dt['qtd_produtos'] > 5, 'A) MAIOR QUE 5'
                              ,np.where(dt['qtd_produtos'] > 3, 'B) MAIOR QUE 3'
                                 ,np.where(dt['qtd_produtos'] > 1, 'C) MAIOR QUE 1'
                                    ,'= 0'        
                              ))) 




display(dt)
```

```python
DT2 = dt.groupby('FX_QTD_PRODUTOS').agg(QTD=('sexo', 'size')
                                        ,SOMA_ACUMULADOS=('dependentes', 'sum')
                                        )
display(DT2)
```

```python
# Criar DataFrame de exemplo
data = {'tipo_pessoa': ['Física', 'Jurídica', 'Física', 'Jurídica', 'Física'],
        'valor': [100, 150, 200, 120, 180]}
df = pd.DataFrame(data)

# Exemplo de DataFrame:
#   tipo_pessoa  valor
# 0      Física    100
# 1    Jurídica    150
# 2      Física    200
# 3    Jurídica    120
# 4      Física    180

# Realizar operação similar ao SQL
resultado = df.groupby('tipo_pessoa').agg(qtd=('tipo_pessoa', 'size')
                                           ,vl_soma=('valor', 'sum')
                                           ,media_soma=('valor', 'mean'))

# Renomear as colunas
resultado = resultado.reset_index()

# Exibir o resultado
display(resultado)
```
