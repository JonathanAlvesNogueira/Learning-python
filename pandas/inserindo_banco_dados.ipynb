```python
import pandas as pd
import numpy as np 

from sqlalchemy import create_engine
```

```python
url = 'D:/donwloads/archive/player_data.csv'
df = pd.read_csv(url)

display(df)



usuario = 'root'
senha = '' # coloque sua senha aqui 
host = 'localhost'
porta = '3306'
banco_de_dados = 'dados_prontos'

# # Criar uma conexÃ£o com o banco de dados MySQL usando SQLAlchemy
engine = create_engine(f'mysql+mysqlconnector://{usuario}:{senha}@{host}:{porta}/{banco_de_dados}')

# Nome da tabela no banco de dados MySQL
tabela_mysql = 'JOGADORES'

# Transferir dados para o banco de dados MySQL
df.to_sql(tabela_mysql, engine, index=False, if_exists='append')

# Verificar os resultados
query = f"SELECT * FROM {tabela_mysql} LIMIT 5;"
resultado = pd.read_sql_query(query, engine)
display(resultado)
```
