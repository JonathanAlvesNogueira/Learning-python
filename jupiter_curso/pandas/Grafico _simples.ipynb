```python
#para usar codigo em formato jupiter dar ".ipynb"

# Importa biblioteca de grafico
import matplotlib.pyplot as plt


x = [x for  x in range(1,100,1)]
y = [y**2 for y in x]

# Plota um gr√°fico de linha com os valores de x no eixo x e os valores de y no eixo y
plt.plot(x,y)
#Adiciona nomes para os eixos 
plt.xlabel('x')
plt.ylabel('y')
#Exibe o grafico 
plt.show()
```
