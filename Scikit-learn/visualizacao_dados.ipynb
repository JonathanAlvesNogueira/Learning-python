```python
import pandas as pd
from ucimlrepo import fetch_ucirepo 

# fetch dataset 
adult = fetch_ucirepo(id=2) 

# data (as pandas dataframes) 
X = adult.data.features 
y = adult.data.targets 

# Combina X e y em um único DataFrame
df = pd.DataFrame(X)
df['target'] = y

# Exibe as primeiras linhas do DataFrame
print(df.head())
```
