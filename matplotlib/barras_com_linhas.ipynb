```python
import pandas as pd 
import matplotlib.pyplot as plt

dados = {'nome': ['Jonathan', 'Carlos', 'Pedro'],
         'bruto': [20, 30, 50],
         'porcentagem': [20, 30, 50]}

df = pd.DataFrame(dados)
display(df)

fig, ax = plt.subplots()

# Plotar o gráfico de barras com rótulos de dados
bars = ax.bar(df['nome'], df['bruto'])

# Adicionar rótulos de dados acima das barras
for bar in bars:
    # Pega a altura de onde vou querer colocar o rotulo de dados 
    yval = bar.get_height()
    #  Seleciona a posiçãodde onde sera colocado o rotulo 
    # (o primeiro parametro pega o meio da barra)
    #  o segundo parametro pega a altura onde o rotulo sera colocado
    # o terceiro 
    ax.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), ha='center', va='bottom')

# Adicionar uma linha com as porcentagens
ax.plot(df['nome'], df['porcentagem'], color='red', marker='o', label='Porcentagens')

plt.legend()
plt.xlabel('Nomes Pessoas')
plt.ylabel('Valores e Porcentagens')
plt.title('Gráfico de Barras com Rótulos de Dados e Linha de Porcentagens')
plt.show()
```
