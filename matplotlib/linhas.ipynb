```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
xa = np.linspace(0,10,5)
cont = np.arange(5)
display(xa)

xb = np.linspace(0,12,4)
cont2 = np.arange(1,5)

# cria grafico de linha
plt.plot(xa, cont, label='empresa1')
plt.plot(xb, cont2, label='empresa2')
plt.legend()

#coloca linha de grade
plt.grid()

#coloca labels no eixo x e y
plt.xlabel('Intervalo')
plt.ylabel('Primeiro grafico')
plt.title('Linhas')
```
