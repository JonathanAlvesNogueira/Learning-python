```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

dados = {'nome': ['Jonas', 'Jonathan']
         ,'porcentagens': [10,90] 
         ,'bruto': [10,90]}

df = pd.DataFrame(dados)
```

```python
fig, ax = plt.subplots()

bars = ax.bar(df['nome'], df['bruto'], color='black')
for bar in bars:
    tamanho = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, tamanho, round(tamanho, 2), ha='center', va='bottom')

ax.plot(df['nome'], df['porcentagens'], color='red', marker='o', label='Porcentagens')
plt.legend()
plt.title('Teste')
plt.show()
```
